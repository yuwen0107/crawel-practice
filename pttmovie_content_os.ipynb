```python
import requests
from bs4 import BeautifulSoup
import os

#創資料夾
path = './pttmovie' #在這個目錄底下,創資料夾
if not os.path.exists(path):
    os.mkdir(path)

url = 'https://www.ptt.cc/bbs/movie/index.html'

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}


#進行迴圈
#迴圈在第一次的時候,會變成上一頁的網址
#再改變網址,再進入下一次迴圈
for i in range(0,3):

    res = requests.get(url,headers=headers)

    soup = BeautifulSoup(res.text,'html.parser')

    #print(soup.prettify())

    title = soup.select('div[class="title"]')
    #print(title)
    for t in title:
        try:
            article_title = t.select('a')[0].text
            
            #該篇文章的網址,對這篇網址進行requests
            article_url ='https://www.ptt.cc' +  t.select('a')[0]['href']
            article_req = requests.get(article_url,headers=headers)
            article_soup = BeautifulSoup(article_req.text,'html.parser')
            #第一個[0],從list裡取出來
            article_content = article_soup.select('div[id="main-container"]')[0].text.split('--')[0]
            
            print(article_title)
            print(article_url)
            print(article_content)
            
            #把每篇文章存成文字檔,放在資料夾裡面
            #開一個文字檔
            with open(path + '/%s.txt'%(article_title),'w',encoding='utf-8') as f: #我要在%s創一個檔案名稱
                #要把article_content寫進去
                f.write(article_content)
            
        except:
            print(t)

    last_page_url = 'https://www.ptt.cc' + soup.select('a[class="btn wide"]')[1]['href'] #上一頁按鈕的網址
    #print(last_page_url)
    url = last_page_url #把url的變數,改成上面獲得的網址
```
