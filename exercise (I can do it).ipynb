```python
import requests
from bs4 import BeautifulSoup
import os

path = './exercise'
if not os.path.exists(path):
    os.mkdir(path)
url = 'https://www.ptt.cc/bbs/movie/index.html'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}

for i in range(0,3):
    res = requests.get(url,headers=headers)
    soup = BeautifulSoup(res.text,'html.parser')
    #print(soup.prettify())
    
    title = soup.select('div[class="title"]')
    #print(title)
    
    for t in title:
        try:
            article_title = t.select('a')[0].text
            article_url ='https://www.ptt.cc'+ t.select('a')[0]['href']
            article_res = requests.get(article_url,headers=headers)
            article_soup = BeautifulSoup(article_res.text,'html.parser')
            content = article_soup.select('div[id="main-content"]')[0].text.split('--')[0]
            print(article_title)
            print(article_url)
            print(content)
            
            with open(path + '/{}.txt'.format(article_title),'w',encoding='utf-8') as f:
                f.write(content)
        except:
            print(article_title)
    last_page_url = 'https://www.ptt.cc'+ soup.select('a[class="btn wide"]')[1]['href']
    print(last_page_url)
    url = last_page_url
```
