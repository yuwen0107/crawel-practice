```python
import requests
from bs4 import BeautifulSoup

url = 'https://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}

post_data_str = '''method: search
searchMethod: true
searchTarget: ATM
orgName: 
orgId: 
hid_1: 1
tenderName: 
tenderId: 
tenderStatus: 5,6,20,28
tenderWay: 
awardAnnounceStartDate: 109/08/14
awardAnnounceEndDate: 109/08/14
proctrgCate: 
tenderRange: 
minBudget: 
maxBudget: 
item: 
hid_2: 1
gottenVendorName: 
gottenVendorId: 
hid_3: 1
submitVendorName: 
submitVendorId: 
location: 
execLocationArea: 
priorityCate: 
isReConstruct: 
btnQuery: 查詢'''
#換行符號

#先宣告一個空的字典
data = {}
for r in post_data_str.split('\n'):
    #每個符號都是 冒號 空格
    key = r.split(': ')[0]
    value = r.split(': ')[1]
    data[key] = value
print(data)

#對網頁進行requests
res = requests.post(url,headers=headers,data=data)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.prettify())
```
